# Delta Tables

## Tipos de tabelas:
### Gerenciadas e não gerenciadas


```python
## Lendo de um arquivo CSV

df2 = spark.read.format("csv").option("header","true").load("Files/Raw/Produto.csv")
```

```python
display(df2)
```

### Método 01

saveAsTable()

- Criando uma tabela atraves do Spark

```python
df2.write.format("delta").mode("overwrite").saveAsTable("User_Gerenciada01")
```

### Método 02

- Criando uma tabela pelo método CTAS (CreateTableAsSelect)

```python
df2.createOrReplaceTempView("user_view")
```

```python
df = spark.sql("SELECT * FROM HML_Lakehouse_GL.user_gerenciada01 LIMIT 1000")
display(df)
```

```python
%%sql

SELECT * FROM HML_Lakehouse_GL.user_gerenciada01
```

```python
## Mostrando o conteudo da View

spark.sql("select * from user_view").show()
```

```python
%%sql

select * from user_view;
```

**Criando a tabela com CTAS**

```python
spark.sql("CREATE TABLE IF NOT EXISTS user_gerenciada02 USING DELTA AS SELECT * FROM USER_VIEW")
```

```python
df = spark.sql("SELECT * FROM HML_Lakehouse_GL.user_gerenciada02 LIMIT 1000")
display(df)
```

### Criando tabela com SQL puro

```python
%%sql

CREATE TABLE IF NOT EXISTS user_gerenciada03
USING DELTA AS 
SELECT * FROM USER_VIEW
```

```python
df = spark.sql("SELECT * FROM HML_Lakehouse_GL.user_gerenciada03 LIMIT 1000")
display(df)
```

### Método 03
**Criando uma tabela vazia usando SQL**

```python
%%sql

CREATE TABLE user_gerenciada04(
    ID INT,
    NOME STRING,
    SOBRENOME STRING,
    EMAIL STRING,
    SEXO STRING,
    DEPARTAMENTO STRING
) USING DELTA
```

```python
%%sql

INSERT INTO user_gerenciada04 (ID, NOME, SOBRENOME, EMAIL, SEXO, DEPARTAMENTO)
VALUES (100, 'Gustavo', 'Lima', 'gustavo.lima@redeoba.com.br', 'Homem', 'TI')
```

```python
df = spark.sql("SELECT * FROM HML_Lakehouse_GL.user_gerenciada04 LIMIT 1000")
display(df)
```

### Inserindo dados da View na tabela

```python
%%sql
insert into user_gerenciada04
select * from user_view
```

## Método 04
**Utilizando o Método Save**
Use o metodo sabeAsTable quando você quiser transformar um Dataframa em uma tabela - Local Padrão
Use o Medodo sabve quando você quiser especificar o local da tabela

```python
## df2 = spark.read.format("csv").option("header","true").load("Files/Raw/Produto.csv")

df2.write.format("delta").mode("overwrite").save("Tables/user_gerenciada05")
```

```python
display(df2)
```
